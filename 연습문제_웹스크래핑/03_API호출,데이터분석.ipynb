3-1. 네이버 책 검색 API 호출하기

```python
from dotenv import load_dotenv
import os
load_dotenv()

CLIENT_ID = os.getenv("CLIENT_ID")
print(CLIENT_ID[:4])

CLIENT_SECRET = os.getenv("CLIENT_SECRET")
print(CLIENT_SECRET[:4])
```

```python
import urllib.request
import requests

def search_books(query):   
    # query string 문자열을 dict 선언
    payload = {
    'query':  query, #'파이썬',
    'display': 100,
    'sort': 'sim'
    }
    return payload

headers = {
    'X-Naver-Client-Id': CLIENT_ID,
    'X-Naver-Client-Secret': CLIENT_SECRET
    }
    
url = 'https://openapi.naver.com/v1/search/book.json'
my_payload = search_books('파이썬')
# requests get(url, params, headers) 요청 
res = requests.get(url, params=my_payload, headers=headers)
# json() 함수로 응답 결과 가져오오기 
items_data = res.json()['items']
```

1. 검색어로 찾은 책 목록을 json 파일로 저장하기

```python
#data/books.json 파일 생성기
import json

with open('data/books.json','w',encoding='utf-8') as file:
    json.dump(items_data, file)
```

2. books.json 파일을 Pandas DataFrame로 저장하기

```python
import pandas as pd

data = pd.read_json('data/books.json')
print(data.shape)
```

3. 검색어로 찾은 책 목록 출력하기

```python
data
```

4. 검색어로 찾은 책 목록 중에서 가격이 2만원 이상인 책만 출력하기

```python
data2 = data[['title','author','discount','publisher','pubdate']]\
.sort_values(by='discount',ascending=False).reset_index(drop=True)

discount_20000 = data2.loc[data2['discount'] >= 20000].copy()
discount_20000
```
